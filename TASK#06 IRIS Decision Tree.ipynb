### <b>Importing Libraries</b>

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_palette('husl')
import graphviz

import matplotlib.pyplot as plt
%matplotlib inline
import pydot
from IPython.display import Image

from sklearn.tree import DecisionTreeClassifier,export_graphviz,plot_tree
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm
```

### <b>Reading the Data.</b>

```python
data = pd.read_csv("Iris.csv")
```

```python
data.head()
```

<b> Dropping 'Id' Column as it is not necessary </b>

```python
data.drop('Id',axis=1,inplace=True)
```

#### <b>Pair-Plotting different combinations from data</b>

```python
sns.pairplot(data,hue='Species',height=3)
plt.show()
```

### <b>Plotting SepalLength v/s SepalWidth and PetalLength v/s PetalWidth</b>

```python
sns.FacetGrid(data,hue='Species').map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
plt.show()
sns.FacetGrid(data,hue='Species').map(plt.scatter,'PetalLengthCm','PetalWidthCm').add_legend()
plt.show()
```

#### <b>Storing data and target in X and y respectively</b>

```python
X = data.iloc[:,:-1]
y = data['Species']
```

#### <b>Training the model </b>

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
tree_obj = DecisionTreeClassifier()
tree_obj.fit(X_train,y_train)
print("Training Complete !")
y_pred = tree_obj.predict(X_test)
```

#### <b>Comparing Actual and Predicted Value</b>

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

#### <b>Decision Tree Graph</b>

```python
features=['Sepal Length','Sepal Width','Petal Length','Petal Width']
class_=['Setosa', 'Versicolor', 'Virginica']

fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=350)
plot_tree(tree_obj,feature_names=features,class_names=class_,filled=True,rounded=True)
```

#### <b>Prediction with random values</b>

```python
print('Class Names : ',class_,"\n")
print('Enter the following values :')
sepal_length=float(input('Enter Sepal Length : '))
sepal_width =float(input('Enter Sepal Width  : '))
petal_length=float(input('Enter Petal Length : '))
petal_width =float(input('Enter Petal Width  : '))
print(f'\nEstimating the Class Probalicties.....')
output = tree_obj.predict([[sepal_length, sepal_width,petal_length,petal_width]])
# output = tree_obj.predict([[5.1, 3.4,1.4,0.2]])
print('Output : ',output[0])
```

#### <b>Accuracy of Model</b>

```python
# Model Accuracy
print('Accuracy : ',sm.accuracy_score(y_test,y_pred))
```

#### <b>Accuracy : 95%</b>
